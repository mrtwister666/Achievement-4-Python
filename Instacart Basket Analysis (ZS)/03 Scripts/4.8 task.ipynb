Question 4

```python
# q4 - create loyalty flags using loc() and transform()

# create new column to contain max orders per user
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check output
ords_prods_merge.head(100)
```

```python
# Q4 loc commands
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

Question 5

```python
# Group by "prices" 
ords_prods_merge.groupby('prices')
```

```python
# Split the data into groups by "loyalty_flag" and then calculate the mean for each "prices" column
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

```python
# loyal customers buy on average 2-3$ less per purchase on average
```

Question 6

```python
# create new average price column
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check output
ords_prods_merge.head(100)
```

```python
# create price flags
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spend_flag'] = 'low spender'
ords_prods_merge.loc[ords_prods_merge['average_price'] > 10, 'spend_flag'] = 'high spender'

```

```python
#check output
ords_prods_merge.head(100)
```

Question 7

```python
# create new freuency column
ords_prods_merge['purchase_frequency'] = ords_prods_merge.groupby(['user_id']).agg({'days_since_prior_order':['mean']})
```

```python
ords_prods_merge.loc[ords_prods_merge['purchase_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'

```

```python
ords_prods_merge.loc[ords_prods_merge['purchase_frequency'] > 10 & (ords_prods_merge['purchase_frequency'] <= 20), 'frequency_flag'] = 'high spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['purchase_frequency'] <= 10, 'frequency_flag'] = 'Non-frequent customer'
```
